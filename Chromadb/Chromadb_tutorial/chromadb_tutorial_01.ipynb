```python
import chromadb
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
```

```python
# Specify the directory to store the database files
path_chromadb_dump = '../Chromad_tutorial/'
persist_dir = '../Chromadb_tutorial/collections/'
```

```python
is_persistent = True
embedding_function = SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")

# Create a Chroma client
client = chromadb.Client(Settings(is_persistent=is_persistent, persist_directory=persist_dir))  

# Create a collection
collection_name = "my_collection"
collection = client.get_or_create_collection(name=collection_name, embedding_function=embedding_function)
```

```python
# Prepare documents
documents = [
  {"id": "doc1", "text": "This is the first document.", "metadata": {"source": "source_1", "key": "1"}},
  {"id": "doc2", "text": "This is the second document, with some more text.", "metadata": {"source": "source_2"}},
  {"id": "doc3", "text": "This is the third document, with some more text.", "metadata": {"source": "source_3", "key": "3"}}    
]
```

```python
# Add documents to the collection
collection.add(
  ids=[doc["id"] for doc in documents],
  documents=[doc["text"] for doc in documents],
  metadatas=[doc["metadata"] for doc in documents]
)
```

```python
# Prepare a query
query = "Find the first document."
n_results = 3
include = ['documents', 'embeddings', 'metadatas']

# results = collection.query(query_texts=query, n_results=3, include=['documents', 'embeddings'])
results = collection.query(query_texts=query, n_results=n_results, include=include)

# Print the results
for result in results["ids"]:
    print(result)
```
