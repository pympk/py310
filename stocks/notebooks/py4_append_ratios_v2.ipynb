```python
import sys
from pathlib import Path

# Notebook cell
%load_ext autoreload
%autoreload 2

# Get root directory (assuming notebook is in root/notebooks/)
NOTEBOOK_DIR = Path.cwd()
ROOT_DIR = NOTEBOOK_DIR.parent if NOTEBOOK_DIR.name == 'notebooks' else NOTEBOOK_DIR

# Add src directory to Python path
sys.path.append(str(ROOT_DIR / 'src'))

# Verify path
print(f"Python will look in these locations:\n{sys.path}")


# --- Execute the processor ---
import utils

SOURCE_PATH, DEST_PATH = utils.main_processor(
    data_dir='..\data',  # search project ..\data
    downloads_dir='',  # None searchs Downloads dir, '' omits search
    downloads_limit=0,  # search the first 10 files
    clean_name_override='df_finviz_n_ratios.pkl',  # override filename
    start_file_pattern='df_finviz.pkl', # search for files starting with 'df_'
)
```

```python
import pandas as pd

# Set pandas display options to show more columns and rows
pd.set_option('display.max_columns', None)  # Show all columns
# pd.set_option('display.max_rows', 10)       # Limit to 10 rows for readability
# pd.set_option('display.width', None)        # Let the display adjust to the window
# pd.set_option('display.max_colwidth', None) # Show full content of each cell
```

```python
# Load the pickle file into a DataFrame
df_finviz = pd.read_pickle(SOURCE_PATH)

# Display basic information about the DataFrame
print(f"df_finviz.shape: {df_finviz.shape}")
print("\nFirst few rows:")
df_finviz.head()
```

```python
# Remove duplicates from df_finviz, keeping the first occurrence
df_finviz = df_finviz[~df_finviz.index.duplicated(keep='first')]

# Verify no duplicates remain
print(f"Duplicates removed. New shape: {df_finviz.shape}")
print(f"Any duplicates remaining? {df_finviz.index.has_duplicates}")
```

```python
# Load the pickle file into a DataFrame
df_perf_ratios= pd.read_pickle('..\data\df_perf_ratios.pkl')

# Display basic information about the DataFrame
print(f"df_perf_ratios.shape: {df_perf_ratios.shape}")
print("\nFirst few rows:")
df_perf_ratios.head()
```

```python
df_perf_ratios.index
```

```python
df_finviz.index
```

```python
# Find items in df_finviz.index that are not in df_perf_ratios.index
missing_tickers = df_finviz.index[~df_finviz.index.isin(df_perf_ratios.index)]
print("Tickers in df_finviz but not in df_perf_ratios:")
print(missing_tickers)
print(f"\nTotal missing tickers: {len(missing_tickers)}")
```

```python
# Remove rows where the index matches missing_tickers
df_finviz = df_finviz.drop(missing_tickers)

# Verify the removal
print(f"Original shape: {df_finviz.shape[0] + len(missing_tickers)}")
print(f"New shape after removing {len(missing_tickers)} missing tickers: {df_finviz.shape}")
```

```python
# Remove duplicates from df_perf_ratios, keeping the first occurrence
df_perf_ratios = df_perf_ratios[~df_perf_ratios.index.duplicated(keep='first')]

# Verify no duplicates remain
print(f"Duplicates removed. New shape: {df_perf_ratios.shape}")
print(f"Any duplicates remaining? {df_perf_ratios.index.has_duplicates}")
```

```python
# combine columns from df_perf_ratios into df_finviz using their indexes as the key
# If columns overlap, use suffixes (e.g., lsuffix='_finviz', rsuffix='_perf').
_df = df_finviz.join(df_perf_ratios, lsuffix='_finviz', rsuffix='_perf')

# Print info about the merged dataframe
print(f"Original df_finviz shape: {df_finviz.shape}")
print(f"Original df_perf_ratios shape: {df_perf_ratios.shape}")
print(f"Merged dataframe shape: {_df.shape}")
```

```python
_df.to_pickle(DEST_PATH)

print(f"Merged dataframe saved to {DEST_PATH}")
```
